```python
import numpy as np 
import sklearn 
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_circles
X,y = make_circles(90, factor=0.2, noise=0.1) 
#noise = standard deviation of Gaussian noise added in data. 
#factor = scale factor between the two circles
plt.scatter(X[:,0],X[:,1], c=y, s=50, cmap='seismic')
plt.show()
```

```python
from sklearn import svm
#clf = svm.SVC()
clf = svm.SVC(kernel='linear')

clf.fit(X, y)

#print(X)
res = clf.predict(X)
#plt.scatter(res[:,0],y[:,0])
#plt.show()
print(res)
print(y)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y, res))
```
